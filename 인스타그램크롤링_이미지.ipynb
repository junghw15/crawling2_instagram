```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from urllib import parse
import requests
from urllib.request import urlopen
import urllib.request 

#인스타그램 열고, 로그인, 검색어 입력

#크롬 브라우저 열고 인스타그램 연결
driver=webdriver.Chrome(executable_path='C:/Users/User/Desktop/크롤링/chromedriver.exe')
driver.get('http://www.instagram.com')
time.sleep(3)

#인스타그램 로그인 정보
id='hw_crawling2' #로그인아이디

input_id=driver.find_elements(By.CSS_SELECTOR,'input._aa4b._add6._ac4d')[0] #아이디 태그 불러오기
#print(input_id)
input_id.clear()
input_id.send_keys(id) #아이디 입력

password= '비밀번호'#로그인 비번
input_pw=driver.find_elements(By.CSS_SELECTOR,'input._aa4b._add6._ac4d')[1] #비밀번호 태그 불러오기
input_pw.clear()
input_pw.send_keys(password) #비밀번호 입력
input_pw.submit()

time.sleep(5)

#검색어 조건에 따른 url 생성 함수
def insta_searching(word):
    url='https://www.instagram.com/explore/tags/'+str(word)
    return url

#게시물 검색 키워드 입력
word=str('배냇저고리')
url=insta_searching(word)

#검색 결과 페이지 열기
driver.get(url)
time.sleep(10)

#이미지 다운로드 함수 생성

def get_image(driver):
    
    image_tags=driver.find_elements(By.CSS_SELECTOR, 'div._aagu._aato div._aagv img') #이미지 태그 불러오기

    url_list=[]
    image_null=driver.find_elements(By.CSS_SELECTOR,'div._9zm2') #사진 여러장인지 확인(다음 화살표 있는지 확인)
    
    i=0
    while len(image_null) !=0: #사진 여러장인 경우(다음 화살표 있는 경우)
            
        image_tags=driver.find_elements(By.CSS_SELECTOR, 'div._aagu._aato div._aagv img')
        url=image_tags[i].get_attribute('src')   
        url_list.append(url)
        driver.find_element(By.CSS_SELECTOR,'div._9zm2').click() #화살표 클릭해서 사진 넘어가게 함
        time.sleep(8)
        image_null=driver.find_elements(By.CSS_SELECTOR,'div._9zm2')
        i+=1
    
    url=image_tags[i].get_attribute('src')  #이미지 url 불러오기
    url_list.append(url)
    
    return url_list

#다음 게시물 클릭하는 함수 생성
def move_next(driver):
    right=driver.find_element(By.CSS_SELECTOR,'div._aaqg._aaqh')
    right.click()
    time.sleep(3)

#첫번째 게시물 클릭
driver.find_element(By.CSS_SELECTOR,'div._aaq8 div._aagw').click()
time.sleep(3)

#이미지 수집 시작
images=[]

target=10 #수집할 게시물 수
for i in range(target):
    try:
        image=get_image(driver) #이미지 가져오기
        images.append(image)
        move_next(driver)
    except:
        time.sleep(2)
        move_next(driver)
    time.sleep(5)
    
#print(images)

#이미지 폴더에 이미지 저장
for i in range(len(images)):
    for j in range(len(images[i])):
        url=images[i][j]
        path='C:/Users/User/Desktop/크롤링/인스타그램_이미지/{}_{}.png'.format(i+1,j+1) #게시물, 사진 순서대로 번호매겨서 이미지 저장
        urllib.request.urlretrieve(url,path) #이미지 파일 생성
        
        print(images[i][j])
```
