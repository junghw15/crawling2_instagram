```python
#id, 본문내용, 해시태그, 좋아요수 출력하는 함수
def get_content(driver):
    
    #본문 내용 출력
    content=driver.find_elements(By.CSS_SELECTOR,'div._ae5q h1._aacl')[0].text

    #id 출력
    id=driver.find_elements(By.CSS_SELECTOR,'div._aaqy a.x1i10hfl')[0].text
    
    #해시태그 출력
    try:
        hash_tag = re.findall('#([0-9a-zA-Z가-힣]*)',content)
        hash_tag=", ".join(hash_tag)  
    except:
        hash_tag=' '
    
    #좋아요수 출력 
    try:
        like=driver.find_elements(By.CSS_SELECTOR,'span a span.x193iq5w span')[0].text #좋아요수 불러오기
    except:
        like='여러 명'  #좋아요수 안나오는 경우 여러명 출력
    
    data=[content, id, hash_tag, like]

    return data



#댓글 출력 함수 생성 (plus 버튼 누르고, 댓글 출력하기)
def get_comment(driver):
    
    # + 버튼 있는지 확인
    plus_null=driver.find_elements(By.CSS_SELECTOR,'div._ae1k div._ae2s div ul._a9z6 li div.x9f619 button._abl-') 
    
    # + 버튼 있으면 다 누르기
    i=0
    while len(plus_null) !=0 and i<len(plus_null): 
        try:
            #플러스 버튼 클릭
            driver.find_elements(By.CSS_SELECTOR, 'div._ae1k div._ae2s div ul li div.x9f619 button._abl- svg.x1lliihq')[i].click()
            time.sleep(5)
            
            #신고창 뜨는 경우 취소 클릭
            button1=driver.find_elements(By.CSS_SELECTOR, 'div.x7r02ix div div._a9-v div._a9-z button._a9--')
            if len(button1)!=0:
                driver.find_elements(By.CSS_SELECTOR, 'div.x7r02ix div div._a9-v div._a9-z button._a9--')[1].click()
                time.sleep(3)
        except:
            pass    
        
        i+=1
        
        #남은 플러스 버튼 개수 업데이트
        plus_null=driver.find_elements(By.CSS_SELECTOR,'div._ae1k div._ae2s div ul._a9z6 li div.x9f619 button._abl- svg.x1lliihq') # +버튼 있는지 확인
        
    #댓글 출력
    comment_id_list=[]
    comment_list=[]
    comment=driver.find_elements(By.CSS_SELECTOR, 'div._a9zr div._a9zs')

    if len(comment) !=0:
        try:

            #댓글 아이디 출력
            for i in range(2, len(comment)+1): 
                comment_id=driver.find_elements(By.CSS_SELECTOR, 'div.x9f619 div.xt0psk2 span._aap6')[i].text
                comment_id_list.append(comment_id)

            #댓글 내용 출력
            for i in range(1, len(comment)): 
                comment_i=driver.find_elements(By.CSS_SELECTOR, 'div._a9zr div._a9zs')[i].text
                comment_list.append(comment_i)
    
        except:
            pass

    #댓글 아이디, 댓글내용 저장
    comment_data=[comment_id_list, comment_list]
    
    return comment_data

#다음 게시물 클릭하는 함수
def move_next(driver):
    right=driver.find_element(By.CSS_SELECTOR,'div._aaqg._aaqh')
    right.click()
    time.sleep(3)


#첫번째 게시물 클릭
driver.find_element(By.CSS_SELECTOR,'div._aaq8 div._aagw').click()
time.sleep(3)


#데이터 수집 시작
results=[]
results_comment=[]

target=5 #수집할 게시물 수
for i in range(target):
    try:
        data=get_content(driver) #본문내용 가져오기
        results.append(data)
        
        data_comment=get_comment(driver) #댓글 가져오기
        results_comment.append(data_comment)
        move_next(driver)
        
    except:
        time.sleep(2)
        move_next(driver)
        
    time.sleep(5)
 
 
#게시물 컬럼별로 리스트에 저장
result_id=[] #아이디
result_content=[] #본문 내용
result_tags=[] #해시태그
result_likes=[] #좋아요
result_content_number1=[] #게시물번호

for i in range(len(results)):
    result_content.append(results[i][0])
    
for i in range(len(results)):
    result_id.append(results[i][1])

for i in range(len(results)):
    result_tags.append(results[i][2])

for i in range(len(results)):
    result_likes.append(results[i][3])

for i in range(target):
    result_content_number1.append(i+1) 
    

#게시물 번호 
for i in range(target):
    try:
        
        length=len(results_comment[i][0])
        if length ==0:
            results_comment[i].append([i+1])
        else:
            results_comment[i].append([i+1]*length)   
        
    except:
        pass
        
#print(results_comment)

#댓글 컬럼별로 리스트에 저장
result_content_number=[] #게시물 번호
result_comment_id=[] #댓글단 사람 아이디
result_comment=[] #댓글

for i in range(len(results_comment)):
    
    #게시물 번호
    a=results_comment[i][2]
    for j in range(len(a)):
        result_content_number.append(a[j])

    #댓글아이디
    b=results_comment[i][0]
    if len(b)==0:
        result_comment_id.append(' ')
    else: 
        result_comment_id +=b
    
    #댓글내용
    c=results_comment[i][1]
    if len(c)==0:
        result_comment.append(' ')
    else: 
        result_comment +=c


#게시물 데이터 프레임 생성
import pandas as pd
import unicodedata
df=pd.DataFrame(list(zip(result_content_number1,result_id,result_content,result_tags,result_likes)),
                columns=['게시물번호','아이디','본문내용','해시태그','좋아요수'])       
df['본문내용']=df['본문내용'].map(lambda x: unicodedata.normalize('NFC', x)) #한글 깨져서 인코딩  


#댓글 데이터 프레임 생성
import pandas as pd
df_comment=pd.DataFrame(list(zip(result_content_number, result_comment_id, result_comment)),
                columns=['게시물번호','댓글아이디','댓글내용'])       

#print(df_comment)


#데이터 프레임 합치기
df_final= pd.merge(df,df_comment, on='게시물번호', how='outer')
print(df_final)

#엑셀파일 생성
#df.to_excel('df.xlsx')
df_comment.to_excel('df_comment3.xlsx')

#df_final.to_excel('df_final.xlsx',index=False)
```
